### Q1: What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make inferences or decisions about a population based on sample data. It involves making an assumption (the null hypothesis) and then testing whether the sample data provides enough evidence to reject it in favor of an alternative hypothesis.

### Q2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
- Null Hypothesis (H₀): States there is no effect or no difference. Example: 'The average weight of apples = 100g'.
- Alternative Hypothesis (H₁): States there is an effect or difference. Example: 'The average weight of apples ≠ 100g'.

Difference: H₀ is a default assumption, while H₁ is tested against it.

### Q3:  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- Significance level (α) is the threshold for deciding whether to reject H₀.  
- Common values: 0.05 or 0.01.  
- Example: α = 0.05 means there’s a 5% risk of rejecting H₀ when it is actually true.  
It helps control the probability of a **Type I error**.

### Q4: What are Type I and Type II errors? Give examples of each.
- Type I Error (False Positive): Rejecting H₀ when it is true.
  Example: Concluding a medicine works when it actually doesn’t.

- Type II Error (False Negative): Failing to reject H₀ when it is false.
  Example: Concluding a medicine doesn’t work when it actually does.

### Q5: What is the difference between a Z-test and a T-test? Explain when to use each.
- Z-test: Used when population variance is known and sample size is large (n > 30).
- T-test: Used when population variance is unknown and sample size is small (n < 30).

Example:
- Z-test → Testing mean height of 500 people.
- T-test → Testing mean height of 20 people.

### Q6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

n, p = 10, 0.5
size = 1000
data = np.random.binomial(n, p, size)

plt.hist(data, bins=range(n+2), edgecolor="black", alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

### Q7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

```python
import numpy as np
from scipy.stats import norm

# Data
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu_0 = 50  # Hypothesized mean

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z = (mean - mu_0) / (std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

print("Mean =", mean)
print("Z =", z)
print("p-value =", p)
print("Conclusion:", "Reject H0" if p < 0.05 else "Fail to Reject H0")
```

### Q8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = np.random.normal(loc=50, scale=5, size=1000)

mean = np.mean(data)
conf_int = stats.norm.interval(0.95, loc=mean, scale=np.std(data, ddof=1)/np.sqrt(len(data)))

print("Mean:", mean)
print("95% Confidence Interval:", conf_int)

plt.hist(data, bins=30, edgecolor="black", alpha=0.7)
plt.axvline(conf_int[0], color="red", linestyle="dashed", label="95% CI Lower")
plt.axvline(conf_int[1], color="green", linestyle="dashed", label="95% CI Upper")
plt.axvline(mean, color="blue", linestyle="solid", label="Mean")
plt.legend()
plt.title("Normal Distribution with 95% CI")
plt.show()
```

### Q9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = np.array([45, 50, 55, 60, 65, 70, 75])
z_scores = stats.zscore(data)

print("Original Data:", data)
print("Z-scores:", z_scores)

plt.hist(z_scores, bins=10, edgecolor="black", alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
